```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,accuracy_score
from sklearn import svm
# Random forest
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, ConfusionMatrixDisplay, precision_score, recall_score, roc_auc_score


import pickle
```

# Random Forest

## Load Dataset

```python
with open("preprocessed_data/train_df.pkl", 'rb') as f:
    train_df = pickle.load(f)
    # Shuffling the training data so that it is not in blocks
    train_df = train_df.sample(frac=1, random_state=42).reset_index(drop=True)
    X_train = train_df.drop(columns=["label"])
    y_train = np.array(train_df[["label"]]).ravel()

with open("preprocessed_data/test_df.pkl", 'rb') as f:
    test_df = pickle.load(f)
    test_df = test_df.sample(frac=1, random_state=42).reset_index(drop=True)
    X_test = test_df.drop(columns=["label"])
    y_test = np.array(test_df[["label"]]).ravel()

X_tr, X_val, y_tr, y_val = train_test_split(X_train, y_train, random_state=42, test_size=0.2)
```

```python
X_tr
```

```python
y_tr
```

## Random Forest (without HOG Features)

```python
rfc = RandomForestClassifier(random_state = 42) 
print("Current params: ", rfc.get_params())

rfc.fit(X_tr, y_tr)

# Predict
y_val_pred = rfc.predict(X_val)

# Print result
print("Accuracy: "+str(accuracy_score(y_val, y_val_pred)))
print('\n')
print(classification_report(y_val, y_val_pred, zero_division = 0))
labels_dict = {0: "tops", 1: "bottoms", 2: "bags", 3: "shoes", 4: "others"}
```

```python
y_pred = rfc.predict(X_test)

print("Accuracy: "+str(accuracy_score(y_test, y_pred)))
print('\n')
print(classification_report(y_test, y_pred, zero_division = 0))
```

## Random Forest (with HOG Features)

```python
with open("preprocessed_data/hog_train_Df.pkl", 'rb') as f:
    train_df_hog = pickle.load(f)
    train_df_hog = train_df_hog.sample(frac=1, random_state=42).reset_index(drop=True)
    X_train_hog = train_df_hog.drop(columns=["label"])
    y_train_hog = np.array(train_df_hog[["label"]]).ravel()

with open("preprocessed_data/hog_test_df.pkl", 'rb') as f:
    test_df_hog = pickle.load(f)
    test_df_hog = test_df_hog.sample(frac=1, random_state=42).reset_index(drop=True)
    X_test_hog = test_df_hog.drop(columns=["label"])
    y_test_hog = np.array(test_df_hog[["label"]]).ravel()

X_tr_hog, X_val_hog, y_tr_hog, y_val_hog = train_test_split(X_train_hog, y_train_hog, random_state=42, test_size=0.2)
```

```python
X_tr_hog
```

```python
y_tr_hog
```

```python
rfc_hog = RandomForestClassifier(random_state = 42) 

rfc_hog.fit(X_tr_hog, y_tr_hog)

# Predict
y_val_pred_hog = rfc_hog.predict(X_val_hog)

# Print result
print("Accuracy: "+str(accuracy_score(y_val_hog, y_val_pred_hog)))
print('\n')
print(classification_report(y_val_hog, y_val_pred_hog, zero_division = 0))
```

```python
y_pred_hog = rfc_hog.predict(X_test_hog)
print("Accuracy: "+str(accuracy_score(y_test_hog, y_pred_hog)))
print('\n')
print(classification_report(y_test_hog, y_pred_hog, zero_division = 0))
```
